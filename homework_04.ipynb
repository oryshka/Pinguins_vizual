```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv"
data = pd.read_csv(url)
print(data.info(0))
```

# countplot

```python
ax = sns.countplot(data, x="year", hue='species', palette="Set2")

ax.tick_params(axis='x')

ax.set(xlabel="Year", ylabel="Count", title="Counting of penguins")

ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[2])

legend = plt.legend()
legend.set_bbox_to_anchor((1, 1))  # Расположение легенды в правом верхнем углу


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Подготовка данных
gender_counts = data.groupby(['year', 'sex']).size().unstack()

# Построение lineplot
sns.lineplot(data=gender_counts, markers=True, linewidth=2.5, palette="Set2")

# Настройка заголовка и меток осей
plt.title('Number of Penguins by Gender over Years')
plt.xlabel('Year')
plt.ylabel('Number of Penguins')

# Отображение графика
plt.show()

```

# histogram

```python
ax = sns.histplot(data, x="body_mass_g", hue='species')
ax.set(xlabel="Weight", ylabel="Count", title="Penguin weight")

palette = ["#F72585", "#7209B7", "#3A0CA3", "#4361EE", "#4CC9F0"]
# ax.bar_label(ax.containers[0])
# ax.bar_label(ax.containers[1])
# ax.bar_label(ax.containers[2])
```

# scatterplot

```python
ax = sns.scatterplot(data, x="bill_length_mm", y="flipper_length_mm", hue='species', palette="Set2")
ax.set(xlabel="The length of bills (mm)", ylabel="The length of flippers (mm)", 
       title="Ratio of bill length to flipper length")
```

# boxplot

```python
palette = ["#F72585", "#7209B7", "#3A0CA3", "#4361EE", "#4CC9F0"]

ax = sns.boxplot(data, x="island", y="body_mass_g", hue="species", whis=[3,97], palette="Set2")

ax.set(xlabel="Island", ylabel="The body mass (g)", title="The weight of pinguins on different islands")
```

# catplot

```python
ax = sns.catplot(data, x="year", y="body_mass_g", hue="species", s=18, palette="Set2")
ax.set(xlabel="Year", ylabel="The body mass (g)", title="The weight of penguins in different years")
```
